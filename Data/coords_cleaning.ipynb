```python
import pandas as pd
```

```python
#Read inb the coordinate csv
city_coords_csv = pd.read_csv("../uscities.csv")
city_coords_csv.head()
```

```python
#Take the columns we need and drop that data that is null
city_coords = city_coords_csv[["city", "state_id", "lat", "lng"]].dropna()
city_coords.head()
```

```python
#renaming the columns
city_coords.rename(columns={'state_id':'state', 'city':'city', 'lng':'lon'}, inplace=True)
city_coords
```

```python
#See the columns and data type of the table
city_coords.info()
```

```python
#send csv out to outputs
city_coords.to_csv('..//csv_outputs//city_coords.csv', index=False)
```
