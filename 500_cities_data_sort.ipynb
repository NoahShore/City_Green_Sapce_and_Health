```python
#Dependencies
import pandas as pd
```

```python
#Import the csv files
mental_health_csv = pd.read_csv('a500_cities_data\\500_Cities__Mental_health_not_good_for___14_days_among_adults_aged___18_years_20231012.csv')
obesity_csv = pd.read_csv('a500_cities_data\\500_Cities__Obesity_among_adults_aged___18_years_20231012.csv')
dr_visits_csv = pd.read_csv('a500_cities_data\\500_Cities__Visits_to_doctor_for_routine_checkup_within_the_past_year_among_adults_aged___18_years_20231012.csv')
```

```python
#Cleaning the table down to needed columns
mh_df = mental_health_csv[['CityName', 'StateAbbr', 'Data_Value', 'DataValueTypeID']].dropna()
mh_df = mh_df[mh_df["DataValueTypeID"].isin(["AgeAdjPrv"])]
mh_df = mh_df.drop(columns=['DataValueTypeID'], inplace=False)
mh_df.rename(columns={'StateAbbr':'state', 'CityName':'city', 'Data_Value':'mh_data_value'}, inplace=True)
mh_df
```

```python
#Cleaning the table down to needed columns
obesity_df = obesity_csv[['CityName', 'StateAbbr', 'Data_Value', 'DataValueTypeID']].dropna()
obesity_df = obesity_df[obesity_df["DataValueTypeID"].isin(["AgeAdjPrv"])]
obesity_df = obesity_df.drop(columns=['DataValueTypeID'], inplace=False)
obesity_df.rename(columns={'StateAbbr':'state', 'CityName':'city', 'Data_Value':'ob_data_value'}, inplace=True)
obesity_df
```

```python
#Cleaning the table down to needed columns
dr_visits_df = dr_visits_csv[['CityName', 'StateAbbr', 'Data_Value', 'DataValueTypeID']].dropna()
dr_visits_df = dr_visits_df[dr_visits_df["DataValueTypeID"].isin(["AgeAdjPrv"])]
dr_visits_df = dr_visits_df.drop(columns=['DataValueTypeID'], inplace=False)
dr_visits_df.rename(columns={'StateAbbr':'state', 'CityName':'city', 'Data_Value':'dv_data_value'}, inplace=True)
dr_visits_df
```

```python
#Export the tables to csv files for SQL
mh_df.to_csv('csv_outputs//mental_health.csv', index=False)
obesity_df.to_csv('csv_outputs//obesity.csv', index=False)
dr_visits_df.to_csv('csv_outputs//dr_visits.csv', index=False)
```
